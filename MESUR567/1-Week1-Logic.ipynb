## Logic in Programming

### Why is logic helpful in programming?

## Conditionals

You run conditionals in your mind every single day. For example, you check your phone in the morning and notice that there is rain forecasted for the afternoon. You tell yourself: "If there is rain, then I want to bring umbrella!" Simple, right? 

Let's make that scenario a bit more complicated: you are a doctor, and you are seeing a patient. He has hypertension and diabetes. You know that to treat the hypertension, you give drug X. To treat the diabetes, you give Drug Y. But, if a patient has both, you give Drug Z. How would you organize that in your mind? Do you categorize the patient as having HTN first, then check if they have diabetes. Or, do you have schemas for HTN, diabetes, and both, as three separate "ifs". Let me write this out...

```python
#Scenario 1: Sequential Statements
HTN = False
Diabetes = True
if HTN == True :
    if Diabetes == True:
        print("Prescribe Drug Z")
    else:
        print("Prescribe Drug X")
else:
    if Diabetes == True:
        print("Prescribe Drug Y")
```

```python
#Scenario 2: Independent Schema
HTN = False
Diabetes = True
if HTN == True and Diabetes == True :
        print("Prescribe Drug Z")
if HTN == True and Diabetes == False:
    print("Prescribe Drug X")
if HTN == False and Diabetes == True:
    print("Prescribe Drug Y")
```

These are two ways to get to the same result (more or less). As tasks get more complicated, you will have to think which method is most efficient for you to write. 

**Question to ponder**: Does my code right now account for all the possible scenarios?

#### I love Trees?

```python
from IPython.display import Image
Image("Media/treekey.png")
```

The image above shows a dichotomous key that a tree-lover would use when determining what type of tree the leaf they found belongs to. There are 11 possible types of tree, dependent on 4 variables (texture, leaf shape, leaf position, and leaf type).

Let's break this down...

```python
#Setting up our variables:
bark = "rough"  # smooth, rough, corky, scaling, scored horizontally
leaves = "neither"  # neither, pinnate, lobed
leaf_shape = "pointed oval"  # pointed oval, heart-shaped, broad lanceolate
leaf_position = "opposite"  # opposite, alternate
```

```python
# The conditionals...
if bark == "rough":
    if leaves == "neither":
        if leaf_shape == "pointed oval":
            print("It's an elm!")
        else:
            print("It's a sweet chestnut!")
    elif leaves == "pinnate":
        print("It's an ash!")
    elif leaves == "lobed":
        print("It's an oak!")

elif bark == "corky":
    print("It's an elder!")

elif bark == "scored horizontally":
    print("It's a rowan!")

elif bark == "scaling":
    if leaves == "neither":
        print("It's a birch!")
    else:
        if leaf_position == "opposite":
            print("It's a sycamore!")
        else:
            print("It's a plane!")

elif bark == "smooth":
    if leaves == "pinnate":
        print("It's an ash!")
    else:
        if leaf_shape == "pointed oval":
            print("It's a beech!")
        else:
            print("It's a lime!")
```

Ooooooh weeeee! That was a lot of code, let's walk through how we accomplished that..
The first variable we want to use to narrow down what path of the dichotomous key we go down is the texture of bark. We have five options (smooth, rough, corky, scaling, scored horizontally). For "corky" and "scored horizontally", we already have our answer, elder or rowan..nice! 

For the other three, we have to follow that dichotomous key into each following subpart. For a rough texture, the leaf type (pinnate, lobed, or neither), really helps us narrow down our choices even more. For all our conditions, we need to do the same thing and we can build our code to be as above.

##### How I Would Approach This + Solution

We identified our variables above. Now, the first thing to do is identify what information do I need first to use this flowchart. Once I identify that it is the pH, I set myself up with two conditions: acidemia or alkalemia. In acidemia and alkalemia depending on your pCO2 and bicarb levels, you will either be in respiratory acidosis, respiratory alkalosis, metabolic acidosis, or metabolic alkalosis. 

For those with metabolic acidosis, we can take this one step further. We look at something called the anion gap (AG) to determine potential causes of the metabolic acidosis. A value of True for the AG will indicate a high AG, and lead you to believe MUDPILES as the possible cause. 

```python
#Variables
pH = 7.2
HCO3 = 15
pCO2 = 30
AG = True
```

```python
if pH < 7.35:
    if pCO2 > 44:
        output = "Respiratory Acidosis"
    if HCO3 < 20:
        if AG == True:
            output = "Anion-Gap Metabolic Acidosis"
        else: 
            output = "Non Anion-Gap Metabolic Acidosis"
if pH > 7.45:
    if pCO2 < 36:
        output = "Respiratory Alkalosis"
    if HCO3 > 28:
        output = "Metabolic Alkalosis"

print(output)
```

## Boolean Operators

Boolean operators are a neat way of quickly sorting through data via conditionals. For exampe, with a dataset of grades on a CBSE exam, you can quickly find all the people who were above a certain cutoff. This is a vital tool to perform operations on certain rows, columns, or cells that fit the condition you are curious about. 

We used a bit of boolean operators above with **True** and **False**, but this is using that same logic but a step further. 
Here's the example on the scores:

```python
import numpy as np
# Define the student names and scores
students = np.array(["Rohan", "Cameron", "Aidan", "Jacob", "Sophie", "Adel"])
scores = np.array([65, 40, 70, 30, 75, 45])
cutoff = 50
```

```python
# Create a boolean array where scores > cutoff
scores_above_cutoff = scores > cutoff
print("This is the boolean array:", scores_above_cutoff)
# Index the students array using the boolean array
students_above_cutoff = students[scores_above_cutoff]
print("These students passed:", students_above_cutoff)
```

This was a really simple way for me to use the scores and cutoffs to identify which students passed the exam! This tool is extremely powerful when working with dataframes to sift through data. All the same operations you are learning now with simple arrays will work the same with dataframes...at the end of the day a dataframe is just a bunch of arrays arranged in a certain way.

## Loops (For and While)

For and while loops provide a type of control that can't be achieved with if statements and boolean operators. They allow you to repeatedly perform one action, or a set of actions as long as certain conditions are met. Basically, you can repeat a block of code over and over for a set of data. For example, with the ABG code we wrote earlier, we can adapt that via for loops to run for a set of patients data and not just for one set of values. 

In quick summary: <br>
**For Loops**: Used to iterate over a sequence (like a list, tuple, string, or range). The loop automatically iterates over each item in the sequence. <br>
**While Loops**:
Used to repeatedly execute a block of code as long as a given condition is true.
You need to manage the loop counter or condition manually

#### Let's adapt the ABG example using a for loop

```python
# List of patient data (each patient has pH, pCO2, HCO3, and AG values)
#Three sample patients:
patients = [
    {'pH': 7.30, 'pCO2': 50, 'HCO3': 24, 'AG': False},
    {'pH': 7.28, 'pCO2': 38, 'HCO3': 18, 'AG': True},
    {'pH': 7.48, 'pCO2': 30, 'HCO3': 32, 'AG': False}
]
```

All we are doing is adding one line of code, and now we can process as many patients at a time if we would like! Isn't that pretty powerful?

```python
# Loop through each patient
for i, patient in enumerate(patients): #only line of code added
    pH = patient['pH']
    pCO2 = patient['pCO2']
    HCO3 = patient['HCO3']
    AG = patient['AG']
    output = ""
    # Evaluate acid-base status based on patient values
    if pH < 7.35:
        if pCO2 > 44:
            output = "Respiratory Acidosis"
        if HCO3 < 20:
            if AG:
                output = "Anion-Gap Metabolic Acidosis"
            else:
                output = "Non Anion-Gap Metabolic Acidosis"
    elif pH > 7.45:
        if pCO2 < 36:
            output = "Respiratory Alkalosis"
        if HCO3 > 28:
            output = "Metabolic Alkalosis"

    # Print the result for each patient
    print(f"Patient {i+1}: {output}")
```

In our case, the while loops was used to continuously simulate the changes in heart rate over a 10 second interval. #### Solution

```python
# Initial patient heart rate (bpm)
heart_rate = 85  # Starting heart rate
critical_heart_rate = 120  # Critical level for intervention
low_heart_rate = 50  # Low level for intervention
time = 0  # Starting time
max_time = 10  # Simulate monitoring for 10 time steps
```

```python
# Simulate monitoring patient's heart rate
import random
while time < max_time:
    print(f"Time {time}: Heart Rate = {heart_rate} bpm")

    # Check for abnormal heart rate conditions
    if heart_rate > critical_heart_rate:
        print("Warning: Critical high heart rate! Immediate intervention required.")
        break  # Exit loop if intervention is needed
    elif heart_rate < low_heart_rate:
        print("Warning: Heart rate too low! Immediate intervention required.")
        break  # Exit loop if intervention is needed
    else:
        print("Heart rate is within the normal range.")
    
    # Simulate heart rate changes over time
    heart_rate += random.randint(-50, 50)
    
    # Increment time for each step
    time += 1

print("Monitoring finished.")
```
